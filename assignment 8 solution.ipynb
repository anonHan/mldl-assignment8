In this assignment students have to transform iris data into 3 dimensions
and plot a 3d chart with transformed dimensions and colour each data
point with specific class

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
iris = datasets.load_iris()
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df['species'] = iris.target
```

```python
 df.head()
```

```python
X = df.drop('species',axis=1)
y = df.species
```

```python
# using PCA for reducing the data to 3 dimensions
pca = decomposition.PCA(n_components=3)

# fitting the data to PCA
pca_data = pca.fit_transform(X)

# converting the PCA data to dataframe
pca_df = pd.DataFrame(pca_data,columns=['EigenVector 1','EigenVector 2','EigenVector 3'])
```

```python
pca_df.head()
```

```python
sns.set_style('darkgrid')
fig = plt.figure(figsize=(8,6))
ax = Axes3D(fig)

ax.scatter3D(xs=pca_df['EigenVector 1'],ys=pca_df['EigenVector 2'],zs=150,c=y,s=70,cmap=plt.cm.Set1)
ax.set_xlabel('First EigenVector',fontsize=15,color='red')
ax.set_ylabel('2nd Eigenvector',fontsize=15,color='red')
ax.set_zlabel('3rd Eigenvector',fontsize=15,color='red')
ax.set_title('ScatterPlot of Top 3 Eigenvectors',fontsize=20,color='r')
plt.show()
```
